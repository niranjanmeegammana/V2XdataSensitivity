```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Performance_by_Dataset_Size_and_Model.csv')


# Set font to serif for academic styling
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "Georgia", "DejaVu Serif"]
})
# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Sample Size', y='F1 Score', hue='Type', marker='o')

plt.title('F1 Score vs. Sample Size by Model Type')
plt.xlabel('Sample Size')
plt.ylabel('F1 Score')
plt.grid(True)
plt.legend(title='Model Type')
plt.tight_layout()
# Save at 300 dpi
plt.savefig('f1_score_vs_sample_size.png', dpi=300)
plt.show()
plt.close()

```

CPU & Memory Usage (200K Samples)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your CSV
df = pd.read_csv('Performance_by_Dataset_Size_and_Model.csv')

# Filter for largest dataset size (200K)
df_200k = df[df['Sample Size'] == 200000]

# Set serif font for publication style
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "Georgia", "DejaVu Serif"]
})

# Set up side-by-side bar plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# CPU Usage Plot
sns.barplot(data=df_200k, x='Type', y='CPU Usage (%)', ax=axes[0], hue='Type')
axes[0].set_title('CPU Usage by Model (200K Samples)')
axes[0].set_xlabel('Model Type')
axes[0].set_ylabel('CPU Usage (%)')
axes[0].grid(True, linestyle='--', linewidth=0.5)

# Memory Usage Plot
sns.barplot(data=df_200k, x='Type', y='Memory Usage (MB)', ax=axes[1], hue='Type')
axes[1].set_title('Memory Usage by Model (200K Samples)')
axes[1].set_xlabel('Model Type')
axes[1].set_ylabel('Memory Usage (MB)')
axes[1].grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.savefig('CPU_Memory_Usage.png', dpi=300)
plt.show()
plt.close()
```

Precision vs. Recall by Sample Size

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Performance_by_Dataset_Size_and_Model.csv')

# Melt the DataFrame to long format for Seaborn
melted_df = df.melt(id_vars=["Sample Size", "Type"],
                    value_vars=["Precision", "Recall"],
                    var_name="Metric",
                    value_name="Score")

# Create a FacetGrid bar plot for Precision and Recall
sns.set(style="whitegrid")
g = sns.catplot(data=melted_df,
                x="Sample Size", y="Score", hue="Type", col="Metric",
                kind="bar", height=5, aspect=1.2, palette="Set2")

# Set titles and labels
g.set_titles("{col_name}")
g.set_axis_labels("Sample Size", "Score")
g.set(ylim=(0.45, 0.95))

# Rotate x-axis labels for readability
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)

plt.tight_layout()

plt.savefig('Precision_Recall_Sample_Size.png', dpi=300)
plt.show()
plt.close()
```

Model Size Impact



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('Performance_by_Dataset_Size_and_Model.csv')

# Group by model type and calculate average model size
model_size_df = df.groupby("Type")["Model Size (MB)"].mean().reset_index()

# Set serif font
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "Georgia", "DejaVu Serif"]
})

# Plot bar chart
plt.figure(figsize=(8, 5))
sns.barplot(data=model_size_df, x='Type', y='Model Size (MB)', hue="Type")

plt.title('Average Model Size by Model Type')
plt.xlabel('Model Type')
plt.ylabel('Model Size (MB)')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('Model_Size_Impact.png', dpi=300)

plt.show()
plt.close()
```
